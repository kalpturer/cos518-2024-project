```python
import matplotlib.pyplot as plt
import numpy as np

from aquarel import load_theme

theme = load_theme("scientific").set_font(size=8)
theme.apply()
current_color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
colors3 = current_color_cycle[:3]
colors5 = current_color_cycle[:5]
```

```python
conflict_rate = "0.00"
sleep = "25"

def plot_histogram(input_files):
    datas = []
    filtered_datas = []
    locations = ['North Virginia', 'Tokyo', 'London']
    medians = []
    count = 0
    for i in range(len(input_files)):
        input_file = input_files[i]
        data = []
        with open(input_file, 'r') as f:
            lines = f.readlines()
    
    
        for line in lines[2:]:
            if line[0] == 'R':
                print("Warning: timedout requests in data")
            else:
                data.append(int(line.strip()[:-2]))
                count += 1

        # remove outliers
        threshold = np.percentile(data, 99)
        data = np.array(data)
        filtered_data = data[data <= threshold]
        
        datas.append(data)
        filtered_datas.append(filtered_data)

        print("Median latency in", locations[i], ":", np.median(data))
        medians.append(np.median(data))
        print("outliers within one percent:", data[data > threshold])
    print("Throughput: ", count / 30)

    plt.hist(filtered_datas[0], bins=100)
    plt.hist(filtered_datas[1], bins=100)
    plt.hist(filtered_datas[2], bins=100)
    plt.xlabel('Latency (ms)')
    plt.ylabel('Frequency')
    plt.title(f'Latency Distribution 3 Replicas (conflict rate {conflict_rate}, latency within 99th percentile)')
    plt.grid(True)
    plt.legend(locations)
    theme.apply_transforms()
    plt.savefig(f'3replicas_sleep{sleep}_conflict{conflict_rate}_dist_global.png', bbox_inches='tight')
    plt.show()

    #####################################################################################
    plt.bar(locations, medians, color=colors3)
    plt.xlabel('Locations')
    plt.ylabel('Median latency (ms)')
    plt.title(f'Median latency with 3 Replicas (conflict rate {conflict_rate})')
    theme.apply_transforms()
    plt.savefig(f'3replicas_sleep{sleep}_conflict{conflict_rate}_loc_global.png', bbox_inches='tight')
    plt.show()

input_files = [f'./data/global/results_3_replicas_virginia_tokyo_london/results_{conflict_rate}conflict/exp3_virginia_sleep{{{sleep}}}_time{{30}}.txt',
               f'./data/global/results_3_replicas_virginia_tokyo_london/results_{conflict_rate}conflict/exp3_tokyo_sleep{{{sleep}}}_time{{30}}.txt',
               f'./data/global/results_3_replicas_virginia_tokyo_london/results_{conflict_rate}conflict/exp3_london_sleep{{{sleep}}}_time{{30}}.txt']
plot_histogram(input_files)
```
